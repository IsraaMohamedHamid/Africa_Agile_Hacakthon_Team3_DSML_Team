```python
import random
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from googletrans import Translator
```

```python
# Initialize DialoGPT model and tokenizer
model_name = "EleutherAI/gpt-neo-2.7B"#"facebook/blenderbot-3B"#model_name = "microsoft/DialoGPT-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Initialize the language translator
translator = Translator()

# Define the languages supported by the chatbot
supported_languages = ["en", "yo", "ig", "ha"]
```

```python
# Create the chatbot response function
def get_response(user_input, user_language):
    # Translate the user input to English (the language DialoGPT understands)
    if user_language not in supported_languages:
        response = "Sorry, the language is not supported. Please use one of the following languages: English, Yoruba, Igbo, Hausa."
    else:
        if user_language != "en":
            user_input = translator.translate(user_input, src=user_language, dest="en").text

        # Encode the user input and generate the chatbot response
        input_ids = tokenizer.encode(user_input, return_tensors="pt")
        bot_output = model.generate(input_ids, max_length=150, num_beams=5, early_stopping=True)
        response = tokenizer.decode(bot_output[0], skip_special_tokens=True)

        # Translate the response back to the user's language
        if user_language != "en":
            response = translator.translate(response, src="en", dest=user_language).text

    return response
```

```python
# Main function to interact with the chatbot
def main():
    print("Chatbot: Hi! I'm here to help. Please enter your language: 'en' for English, 'yo' for Yoruba, 'ig' for Igbo, 'ha' for Hausa.")
    user_language = input("You: ").lower()

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye! Take care.")
            break

        response = get_response(user_input, user_language)
        print(f"Chatbot: {response}")
```

```python
if __name__ == "__main__":
    main()
```
